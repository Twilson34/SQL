```python
import pandas as pd
import datetime
```

```python
cars = "../Resources/vehicles.csv"
```

```python
vehicles_df = pd.read_csv(cars)
vehicles_df
```

```python
test=vehicles_df[vehicles_df['manufacturer'].notna()]
price_fix=vehicles_df[vehicles_df['price'] > 0]
new_df=price_fix[price_fix['model'].notna()]
lat_lng_df=new_df[new_df['lat'].notna()]
lat_lng_df=lat_lng_df[lat_lng_df['long'].notna()]

lat_lng_df
```

```python
# I chose to remove rows with na for year (910 cars) since I thought that was key data
no_nas=lat_lng_df[lat_lng_df['year'].notna()]

# I also changed the decimal in the year column
no_nas.year = no_nas.year.apply(int)
no_nas
```

```python
no_nas.count()
```

```python
final_df=no_nas[['manufacturer','model','year','price','image_url','description','paint_color', 'state','region','lat','long']]
final_df.head()
```

```python
final_car=final_df.rename(columns={"manufacturer": "make",
                            "price":"craigslist_price",
                            "region": "city",
                            "long": "lng"})

final_car
```

```python
#Double check which cars columns have NAs
all_na = []
for i in range(8):
    #False treated as 0, true treated as 1
    na_count = final_car.iloc[:,i].isnull().values.sum()
    colname = final_car.columns[i]
    percent_na = round((na_count/len(final_car))*100,2)
    
    col_na = {
        'colname': colname,
        'na_count': na_count,
        'percent_na': percent_na
        }
    
    all_na.append(col_na)
```

```python
#There are some rows with missing makes and missing paint color
all_na
```

```python
#Decided to drop the rows with missing makes, as this is important for data matching
final_car2 = final_car[final_car['make'].notna()]
```

```python
#Checked to see if all na caught above were removed
print(len(final_car))
print(len(final_car2))
```

```python
#write the df to csv for importation into sql
final_car2.to_csv('../Resources/final_car.csv', index=False)
```
