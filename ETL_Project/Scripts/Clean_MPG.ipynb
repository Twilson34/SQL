```python
import pandas as pd
import datetime
```

```python
csv="../Resources/vehicles_mpg.csv"
```

```python
mpg=pd.read_csv(csv)
mpg
```

```python
#Get the column names
list(mpg.columns) 
```

```python
mpg_df = mpg[['make','model','year','comb08','fuelType1','VClass','cylinders','displ','drive','trany','sCharger','tCharger']]
```

```python
# I added super charger and turbo charger; both things that effect fuel economy and price. And of course 
# comb08 which is mpg.
# I took out mpgData and barrels08 which don't tell us anything.  I also took out createdOn and modifiedOn since 
# those were EPA dates.  Also engID (was an EPA designation) and eng_dscr I'm fine with putting these back but they didn't seem useful

mpg_df
```

```python
all_na = []
for i in range(12):
    #False treated as 0, true treated as 1
    na_count = mpg_df.iloc[:,i].isnull().values.sum()
    colname = mpg_df.columns[i]
    percent_na = round((na_count/len(mpg_df))*100,2)
    
    col_na = {
        'colname': colname,
        'na_count': na_count,
        'percent_na': percent_na
        }
    
    all_na.append(col_na)
```

```python
all_na
```

```python
#Rename columns
final_mpg=mpg_df.rename(columns={"comb08": "mpg",
                            "VClass":"size_class",
                            "fuelType1":"fuel",
                            "displ": "displacement_liters",
                            "trany": "transmission",
                            "sCharger": "super_charger",
                            "tCharger": "turbo_charger"})
```

```python
final_mpg['make'] = final_mpg['make'].str.lower()
final_mpg['model'] = final_mpg['model'].str.lower()
final_mpg['fuel'] = final_mpg['fuel'].str.lower()
final_mpg['size_class'] = final_mpg['size_class'].str.lower()
final_mpg['drive'] = final_mpg['drive'].str.lower()
final_mpg['transmission'] = final_mpg['transmission'].str.lower()

final_mpg = final_mpg[final_mpg['cylinders'].notna()]
final_mpg.cylinders = final_mpg.cylinders.apply(int)

final_mpg['turbo_charger'] = final_mpg['turbo_charger'].fillna('FALSE')
final_mpg['super_charger'] = final_mpg['super_charger'].fillna('FALSE')

final_mpg["turbo_charger"].replace({"T": "TRUE"}, inplace=True)
final_mpg["super_charger"].replace({"S": "TRUE"}, inplace=True)
```

```python
#write the df to csv for importation into sql
final_mpg.to_csv('../Resources/final_mpg.csv', index=False)
```
